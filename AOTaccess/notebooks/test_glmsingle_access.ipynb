# Test GLMSingleAccess Class

This notebook tests all functions in the GLMSingleAccess class.

```python
import sys
import numpy as np
from pathlib import Path
from pprint import pprint
sys.path.append('..')
from glmsingle_access import GLMSingleAccess
```

```python
# Initialize the GLMSingleAccess instance
glm_access = GLMSingleAccess()

# Test parameters
test_sub = 1  # test subject
test_ses = 1  # test session
test_video = 1  # test video number
```

## Test Basic Path Functions

```python
# Test get_glm_dir_path
pprint("GLM Main Directory:")
pprint(glm_access.get_glm_dir_path())

# Test get_nii_dir_path
pprint("\nNii Directory for test subject and session:")
pprint(glm_access.get_nii_dir_path(test_sub, test_ses))
```

## Test Data Loading Functions

```python
# Test read_shape
print("Data shape:")
shape = glm_access.read_shape(test_sub, test_ses)
print(shape)
```

```python
# Test read_betas
print("Loading betas:")
betas = glm_access.read_betas(test_sub, test_ses)
print(f"Betas shape: {betas.shape}")
```

```python
# Test read_affine
print("Loading affine matrix:")
affine = glm_access.read_affine(test_sub)
print(f"Affine matrix:\n{affine}")
```

```python
# Test read_header
print("Loading header:")
header = glm_access.read_header(test_sub)
print(f"Header info:\n{header}")
```

```python
# Test read_meanvol
print("Loading mean volume:")
meanvol = glm_access.read_meanvol(test_sub, test_ses)
print(f"Mean volume shape: {meanvol.shape}")
```

```python
# Test read_R2
print("Loading R2:")
R2 = glm_access.read_R2(test_sub, test_ses)
print(f"R2 shape: {R2.shape}")
```

```python
# Test read_R2_mask
print("Creating R2 mask:")
R2_mask = glm_access.read_R2_mask(test_sub, test_ses, threshold=0.2)
print(f"R2 mask shape: {R2_mask.shape}")
print(f"Number of voxels above threshold: {np.sum(R2_mask)}")
```

```python
# Test read_video_betas
print("Loading video betas:")
video_betas = glm_access.read_video_betas(test_sub, test_video)
if video_betas is not None:
    print(f"Video betas shape: {video_betas.shape}")
else:
    print("Video betas file not found")
```

## Test Different Parameters

```python
# Test with different GLM types and resolutions
alternate_glmtype = "TYPED_FITHRF_GLMDENOISE_RR"
alternate_resolution = "1.7mm"

print("Testing with alternate parameters:")
betas = glm_access.read_betas(test_sub, test_ses, 
                             glmtype=alternate_glmtype,
                             resolution=alternate_resolution)
print(f"Betas shape with alternate parameters: {betas.shape}")
```
