# Test ExpLogAccess Methods

This notebook tests all methods in the ExpLogAccess class.

```python
import sys
from pathlib import Path
from pprint import pprint

# Add parent directory to path to import ExpLogAccess
sys.path.append('..')
from explog_access import ExpLogAccess
```

```python
# Initialize ExpLogAccess
exp_log = ExpLogAccess()

# Test parameters
test_sub = 1  # subject 1
test_ses = 1  # session 1
test_run = 1  # run 1
```

```python
# Test get_func_dir
func_dir = exp_log.get_func_dir(test_sub, test_ses)
pprint(f"Functional directory path: {func_dir}")
```

```python
# Test read_events_tsv
try:
    events = exp_log.read_events_tsv(test_sub, test_ses, test_run)
    pprint("Events data:")
    pprint(events[:5])  # Display first 5 events
except FileNotFoundError:
    pprint("Events TSV file not found")
```

```python
# Test read_eyetrack_edf
try:
    eyetrack_data = exp_log.read_eyetrack_edf(test_sub, test_ses, test_run)
    pprint("Eyetracking data:")
    pprint(eyetrack_data)
except FileNotFoundError:
    pprint("Eyetracking EDF file not found")
```

```python
# Test get_frames_pdf
frames_pdf_path = exp_log.get_frames_pdf(test_sub, test_ses, test_run)
pprint(f"Frames PDF path: {frames_pdf_path}")
pprint(f"File exists: {frames_pdf_path.exists()}")
```

```python
# Test read_lot_txt
try:
    log_content = exp_log.read_lot_txt(test_sub, test_ses, test_run)
    pprint("Log content (first 200 characters):")
    pprint(log_content[:200])
except FileNotFoundError:
    pprint("Log TXT file not found")
```

```python
# Test read_scanphyslog_log
try:
    scanphys_content = exp_log.read_scanphyslog_log(test_sub, test_ses, test_run)
    pprint("Scanphyslog content (first 200 characters):")
    pprint(scanphys_content[:200])
except FileNotFoundError:
    pprint("Scanphyslog file not found")
```
