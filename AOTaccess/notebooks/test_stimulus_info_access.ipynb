```python
# Test get_video_path
print("Video path:", sia.get_video_path(test_video_id))

# Test get_video_annotation_dir
print("\nVideo annotation directory:", sia.get_video_annotation_dir(test_video_id))
```

```python
# Test SBERT embeddings
sbert_emb = sia._temp_read_sbert_embeddings(test_video_id)
sbert_pca = sia._temp_read_sbert_embeddings_PCA(test_video_id)
sbert_sae = sia._temp_read_sbert_embeddings_SAE(test_video_id)

print("SBERT embeddings shape:", sbert_emb.shape)
print("SBERT PCA embeddings shape:", sbert_pca.shape)
print("SBERT SAE embeddings shape:", sbert_sae.shape)
```

## Test Video Analysis Methods

```python
# Test semantic segmentation
semantic_seg = sia.read_semantic_segmentation(test_video_id)
print("Semantic segmentation shape:", semantic_seg.shape)

# Test action detection
action_det = sia.read_action_detection(test_video_id)
print("\nAction detection shape:", action_det.shape)

# Test action classification
action_class = sia.read_action_classification(test_video_id)
print("\nAction classification (first 3 rows):", action_class[:3])

# Test captioning
captions = sia.read_captioning(test_video_id)
print("\nCaptions (first 3 rows):", captions[:3])

# Test depth estimation
depth = sia.read_depth_estimation(test_video_id)
print("\nDepth estimation shape:", depth.shape)

# Test keypoint detection
keypoints = sia.read_keypoint_detection(test_video_id)
print("\nKeypoint detection shape:", keypoints.shape)

# Test instance segmentation
instance_seg = sia.read_instance_segmentation(test_video_id)
print("\nInstance segmentation shape:", instance_seg.shape)

# Test object detection
obj_det = sia.read_object_detection(test_video_id)
print("\nObject detection shape:", obj_det.shape)
```
