```python
# Test get_video_path
pprint(f"Video path: {sia.get_video_path(test_video_id)}")

# Test get_video_annotation_dir
pprint(f"\nVideo annotation directory: {sia.get_video_annotation_dir(test_video_id)}")
```

```python
# Test SBERT embeddings
sbert_emb = sia._temp_read_sbert_embeddings(test_video_id)
sbert_pca = sia._temp_read_sbert_embeddings_PCA(test_video_id)
sbert_sae = sia._temp_read_sbert_embeddings_SAE(test_video_id)

pprint(f"SBERT embeddings shape: {sbert_emb.shape}")
pprint(f"SBERT PCA embeddings shape: {sbert_pca.shape}")
pprint(f"SBERT SAE embeddings shape: {sbert_sae.shape}")
```

## Test Motion Energy Features

```python
# Test motion energy features
motion_energy = sia._temp_read_motion_energy_features(test_video_id)
pprint(f"Motion energy features shape: {motion_energy.shape}")
```

## Test Video Analysis Methods

```python
# Test semantic segmentation
semantic_seg = sia.read_semantic_segmentation(test_video_id)
pprint(f"Semantic segmentation shape: {semantic_seg.shape}")

# Test action detection
action_det = sia.read_action_detection(test_video_id)
pprint(f"\nAction detection shape: {action_det.shape}")

# Test action classification
action_class = sia.read_action_classification(test_video_id)
pprint(f"\nAction classification (first 3 rows): {action_class[:3]}")

# Test captioning
captions = sia.read_captioning(test_video_id)
pprint(f"\nCaptions (first 3 rows): {captions[:3]}")

# Test depth estimation
depth = sia.read_depth_estimation(test_video_id)
pprint(f"\nDepth estimation shape: {depth.shape}")

# Test keypoint detection
keypoints = sia.read_keypoint_detection(test_video_id)
pprint(f"\nKeypoint detection shape: {keypoints.shape}")

# Test instance segmentation
instance_seg = sia.read_instance_segmentation(test_video_id)
pprint(f"\nInstance segmentation shape: {instance_seg.shape}")

# Test object detection
obj_det = sia.read_object_detection(test_video_id)
pprint(f"\nObject detection shape: {obj_det.shape}")
```
