# All Function Usage Display

This notebook demonstrates the usage of all functions from various access classes.

```python
import sys
from pathlib import Path
from pprint import pprint
sys.path.append('..')
from AOTaccess.stimulus_info_access import StimuliInfoAccess
from AOTaccess.glmsingle_access import GLMSingleAccess
from AOTaccess.explog_access import ExpLogAccess
from AOTaccess.expdesign_access import ExpDesignAccess
from AOTaccess.bids_access import BidsAccess
```

## StimuliInfoAccess Class

```python
# Initialize StimuliInfoAccess
stimuli_access = StimuliInfoAccess()
test_video_id = 1

# Example usage of StimuliInfoAccess methods
video_path = stimuli_access.get_video_path(test_video_id)
pprint(f"Video path: {video_path}")

llama_description = stimuli_access._temp_read_llama_description(test_video_id)
pprint(f"LLaMA Description v1: {llama_description}")

sbert_embeddings = stimuli_access._temp_read_sbert_embeddings(test_video_id)
pprint(f"SBERT embeddings shape: {sbert_embeddings.shape}")
```

## GLMSingleAccess Class

```python
# Initialize GLMSingleAccess
glm_access = GLMSingleAccess()
test_sub = 1
test_ses = 1

# Example usage of GLMSingleAccess methods
glm_dir_path = glm_access.get_glm_dir_path()
pprint(f"GLM Directory Path: {glm_dir_path}")

betas = glm_access.read_betas(test_sub, test_ses)
pprint(f"Betas shape: {betas.shape}")
```

## ExpLogAccess Class

```python
# Initialize ExpLogAccess
exp_log = ExpLogAccess()
test_sub = 1
test_ses = 1
test_run = 1

# Example usage of ExpLogAccess methods
func_dir = exp_log.get_func_dir(test_sub, test_ses)
pprint(f"Functional directory path: {func_dir}")

try:
    events = exp_log.read_events_tsv(test_sub, test_ses, test_run)
    pprint("Events data:")
    pprint(events[:5])  # Display first 5 events
except FileNotFoundError:
    pprint("Events TSV file not found")
```

## ExpDesignAccess Class

```python
# Initialize ExpDesignAccess
exp_design = ExpDesignAccess()
sub = 1
ses = 1
run = 1

# Example usage of ExpDesignAccess methods
expdesign = exp_design.read_expdesign_file(sub, ses, run)
pprint("Expdesign for single run:")
pprint(expdesign)

session_movies = exp_design.read_session_expdesign(sub, ses)
pprint("Session movies:")
pprint(session_movies)
```

## BidsAccess Class

```python
# Initialize BidsAccess
bids_dir = Path("/research/FGB-CognitivePsychology-Knapen/shared/2024/visual/AOT/bids/aotfull_final")
bids_access = BidsAccess(bids_dir)
test_sub = 1
test_ses = 1
test_run = 1

# Example usage of BidsAccess methods
func_dir = bids_access.get_func_dir(test_sub, test_ses)
pprint(f"Functional directory path: {func_dir}")

try:
    events_data = bids_access.read_events_tsv(test_sub, test_ses, test_run)
    pprint(f"Events TSV Loaded: {len(events_data)}")
except FileNotFoundError:
    pprint("Events TSV file not found")
```
