# Test BidsAccess Class

This notebook tests all functions in the BidsAccess class with a structure similar to test_glmsingle.

```python
import sys
from pathlib import Path
from pprint import pprint
sys.path.append('..')
from AOTaccess.bids_access import BidsAccess

bids_dir = Path("/research/FGB-CognitivePsychology-Knapen/shared/2024/visual/AOT/bids/aotfull_final")
bids_access = BidsAccess(bids_dir)
test_sub = 1
test_ses = 1
test_run = 1
```

## Test Basic Path Functions

```python
print('BIDS Directory:', bids_access.bids_dir)
pprint({'func_dir': bids_access.get_func_dir(test_sub, test_ses)})
pprint({'anat_dir': bids_access.get_anat_dir(test_sub, test_ses)})
pprint({'beh_dir': bids_access.get_beh_dir(test_sub, test_ses)})
pprint({'fmap_dir': bids_access.get_fmap_dir(test_sub, test_ses)})
```

## Test Data Loading Functions

```python
# Example calls demonstrating how each method might be tested
try:
    events_data = bids_access.read_events_tsv(test_sub, test_ses, test_run)
    print('Events TSV Loaded:', len(events_data))
except FileNotFoundError:
    print('Events TSV file not found')

try:
    bold_img = bids_access.read_bold(test_sub, test_ses, test_run)
    print('BOLD image shape:', bold_img.shape)
except FileNotFoundError:
    print('BOLD file not found')

try:
    sbref_img = bids_access.read_sbref(test_sub, test_ses)
    print('SBref image shape:', sbref_img.shape)
except FileNotFoundError:
    print('SBref file not found')

try:
    inplane_img = bids_access.read_inplane_T2(test_sub, test_ses)
    print('Inplane T2 image shape:', inplane_img.shape)
except FileNotFoundError:
    print('Inplane T2 file not found')

try:
    memory_events_data = bids_access.read_memory_events(test_sub, test_ses, test_run)
    print('Memory events loaded:', len(memory_events_data))
except FileNotFoundError:
    print('Memory events file not found')

try:
    fmap_img = bids_access.read_fmap_file(test_sub, test_ses, "MB", "AP", test_run)
    print('Fmap image shape:', fmap_img.shape)
except FileNotFoundError:
    print('Fmap file not found')

try:
    dataset_json = bids_access.read_global_json('dataset_description.json')
    pprint(dataset_json)
except FileNotFoundError:
    print('Global JSON not found')

try:
    acq_json = bids_access.read_acq_json('ex_acq', 1)
    pprint(acq_json)
except FileNotFoundError:
    print('Acquisition JSON not found')

try:
    output_csv_data = bids_access.read_output_csv(test_sub, test_ses)
    print('Output CSV rows:', len(output_csv_data))
except FileNotFoundError:
    print('Output CSV file not found')

try:
    phasediff_img = bids_access.read_phasediff(test_sub, test_ses, test_run)
    print('Phasediff image shape:', phasediff_img.shape)
except FileNotFoundError:
    print('Phasediff file not found')
```
